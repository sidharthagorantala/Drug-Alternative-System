```python
import re

def clean_ingredient(text):
    text = str(text).lower()
    
    # remove percentages like 0.5%
    text = re.sub(r'\d+(\.\d+)?%', '', text)
    
    # remove mg, mcg, g, ml, w/v etc
    text = re.sub(r'\b(mg|mcg|g|ml|w/v|wv)\b', '', text)
    
    # remove numbers
    text = re.sub(r'\d+(\.\d+)?', '', text)
    
    # remove special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    return text.strip()

dataset["clean_ingredient"] = dataset["active_ingredient"].apply(clean_ingredient)

dataset[["active_ingredient", "clean_ingredient"]].head()
```

```python
import re

def clean_ingredient(text):
    text = str(text).lower()
    
    # remove percentage values like 0.5%
    text = re.sub(r'\d+(\.\d+)?%', '', text)
    
    # remove dosage units attached to numbers (e.g., 500mg, 10ml, 5mcg)
    text = re.sub(r'\d+(\.\d+)?\s*(mg|mcg|g|ml)', '', text)
    
    # remove remaining standalone units
    text = re.sub(r'\b(mg|mcg|g|ml|w/v|wv)\b', '', text)
    
    # remove special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # remove extra spaces
    text = re.sub(r'\s+', ' ', text)
    
    return text.strip()

dataset["clean_ingredient"] = dataset["active_ingredient"].apply(clean_ingredient)

dataset[["active_ingredient", "clean_ingredient"]].head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_drug(drug_name, top_n=5):
    
    drug_name = drug_name.lower()
    
    # Check if drug exists
    if drug_name not in dataset["drug_name"].str.lower().values:
        return "Drug not found"
    
    # Get index of drug
    index = dataset[dataset["drug_name"].str.lower() == drug_name].index[0]
    
    # Compute similarity only for that drug
    similarity_scores = cosine_similarity(
        tfidf_matrix[index], tfidf_matrix
    ).flatten()
    
    # Get top similar drugs
    similar_indices = similarity_scores.argsort()[-(top_n+1):-1][::-1]
    
    return dataset.iloc[similar_indices]["drug_name"].values
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_drug_partial(user_input, top_n=5):
    
    user_input = user_input.lower()
    
    # Find drugs containing user input
    matching_indices = dataset[
        dataset["drug_name"].str.lower().str.contains(user_input)
    ].index
    
    if len(matching_indices) == 0:
        return "No matching drug found"
    
    # Take first matching drug as base
    index = matching_indices[0]
    
    base_drug = dataset.iloc[index]["drug_name"]
    base_form = dataset.iloc[index]["dosage_form"]
    
    # Compute similarity
    similarity_scores = cosine_similarity(
        tfidf_matrix[index], tfidf_matrix
    ).flatten()
    
    # Sort by similarity
    similar_indices = similarity_scores.argsort()[::-1]
    
    results = []
    
    for i in similar_indices:
        if dataset.iloc[i]["dosage_form"] == base_form and i != index:
            results.append(dataset.iloc[i]["drug_name"])
        if len(results) == top_n:
            break
    
    return {
        "Selected Drug": base_drug,
        "Dosage Form": base_form,
        "Suggested Alternatives": results
    }
```

```python
recommend_drug_partial("Moxrest")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from rapidfuzz import process

def recommend_advanced(user_input, top_n=5):
    
    user_input = user_input.lower()
    
    # Fuzzy match using clean_drug_name
    best_match = process.extractOne(
        user_input,
        dataset["clean_drug_name"].tolist()
    )
    
    if best_match is None:
        return "Drug not found"
    
    match_clean_name = best_match[0]
    
    matched_rows = dataset[dataset["clean_drug_name"] == match_clean_name]
    
    if matched_rows.empty:
        return "Drug not found"
    
    index = matched_rows.index[0]
    
    selected_drug = dataset.loc[index, "drug_name"]
    base_form = dataset.loc[index, "dosage_form"]
    
    similarity_scores = cosine_similarity(
        tfidf_matrix[index], tfidf_matrix
    ).flatten()
    
    similar_indices = similarity_scores.argsort()[::-1]
    
    results = []
    
    for i in similar_indices:
        if dataset.iloc[i]["dosage_form"] == base_form and i != index:
            results.append({
                "Drug": dataset.iloc[i]["drug_name"],
                "Similarity Score": round(similarity_scores[i], 3)
            })
        if len(results) == top_n:
            break
    
    return {
        "Selected Drug": selected_drug,
        "Dosage Form": base_form,
        "Alternatives": results
    }
```

```python
from rapidfuzz import process

def smart_fuzzy_match(user_input):
    
    user_input = user_input.lower()
    
    # Step 1: Partial match first
    partial_matches = dataset[
        dataset["clean_drug_name"].str.contains(user_input)
    ]
    
    if not partial_matches.empty:
        return partial_matches.iloc[0]["drug_name"]
    
    # Step 2: Fallback fuzzy match
    best_match = process.extractOne(
        user_input,
        dataset["clean_drug_name"].tolist()
    )
    
    if best_match is None:
        return None
    
    matched_name = best_match[0]
    
    return dataset[
        dataset["clean_drug_name"] == matched_name
    ]["drug_name"].values[0]
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

def recommend_advanced(user_input, top_n=5):
    
    selected_drug = smart_fuzzy_match(user_input)
    
    if selected_drug is None:
        return "Drug not found"
    
    matched_rows = dataset[dataset["drug_name"] == selected_drug]
    
    if matched_rows.empty:
        return "Drug not found"
    
    index = matched_rows.index[0]
    
    base_form = dataset.loc[index, "dosage_form"]
    
    similarity_scores = cosine_similarity(
        tfidf_matrix[index], tfidf_matrix
    ).flatten()
    
    similar_indices = similarity_scores.argsort()[::-1]
    
    results = []
    
    for i in similar_indices:
        if dataset.iloc[i]["dosage_form"] == base_form and i != index:
            results.append([
                dataset.iloc[i]["drug_name"],
                round(float(similarity_scores[i]), 3)
            ])
        if len(results) == top_n:
            break
    
    # Convert to DataFrame
    result_df = pd.DataFrame(results, columns=["Alternative Drug", "Similarity Score"])
    
    print("Selected Drug:", selected_drug)
    print("Dosage Form:", base_form)
    
    return result_df
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

def recommend_advanced(user_input, top_n=5):
    
    selected_drug = smart_fuzzy_match(user_input)
    
    if selected_drug is None:
        return "Drug not found"
    
    matched_rows = dataset[dataset["drug_name"] == selected_drug]
    
    if matched_rows.empty:
        return "Drug not found"
    
    index = matched_rows.index[0]
    
    base_form = dataset.loc[index, "dosage_form"]
    base_ingredient = dataset.loc[index, "clean_ingredient"]
    
    similarity_scores = cosine_similarity(
        tfidf_matrix[index], tfidf_matrix
    ).flatten()
    
    similar_indices = similarity_scores.argsort()[::-1]
    
    results = []
    rank = 1
    
    for i in similar_indices:
        if dataset.iloc[i]["dosage_form"] == base_form and i != index:
            
            results.append([
                rank,
                dataset.iloc[i]["drug_name"],
                dataset.iloc[i]["clean_ingredient"],
                round(float(similarity_scores[i] * 100), 2)  # percentage
            ])
            
            rank += 1
            
        if len(results) == top_n:
            break
    
    result_df = pd.DataFrame(
        results,
        columns=[
            "Rank",
            "Alternative Drug",
            "Active Ingredient",
            "Similarity (%)"
        ]
    )
    
    print("Selected Drug:", selected_drug)
    print("Dosage Form:", base_form)
    print("Active Ingredient:", base_ingredient)
    
    return result_df
```
